```python
def dayformat(day):
    maanden = ['januari', 'februari', 'maart', 'april', 'mei', 'juni',
               'juli', 'augustus', 'september', 'oktober', 'november', 'december']
    dag = day.split()
    dag[1] = maanden.index(dag[1]) + 1
    return date(int(dag[2]), dag[1], int(dag[0]))
```
